```python
from ydata_profiling import ProfileReport
import pandas as pd
import pdfkit
```

```python
df = pd.read_csv('AirPassengers.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df.columns = ['Date', 'Passengers']
```

```python
df['Month'] = (df['Date'].str.split('-').str[0]).astype(int)
df['Year'] = (df['Date'].str.split('-').str[1]).astype(int)
```

```python
df.head()
```

```python
df = df[['Year', 'Month', 'Passengers']]
```

```python
df.head()
```

```python
profile = ProfileReport(df, title="AirPassengers Dataset Profiling", explorative=True)

```

```python
html_file = "AirPassengers Dataset Profiling.html"
profile.to_file(html_file)

# Convert the HTML file to a PDF
pdf_file = "AirPassengers Dataset Profiling.pdf"
pdfkit_config = pdfkit.configuration(wkhtmltopdf='/usr/bin/wkhtmltopdf')
pdfkit.from_file(html_file, pdf_file, configuration=pdfkit_config)

print(f"PDF report has been saved as {pdf_file}")
```
